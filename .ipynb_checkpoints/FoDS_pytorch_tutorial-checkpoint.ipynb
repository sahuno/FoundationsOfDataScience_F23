This notebook is a part of an interactive lecture on stochastic gradient
descent (SGD), deep learning, and pytorch. The goal after walking through
this is to have an idea of how to do the following in pytorch:

1) Write down a probabilistic model as a loss function

2) Optimize a loss function with SGD

3) Extend a basic (e.g. linear) model to use a neural network instead

4) Do basic mini-batching and early stopping.

We will use a similar example from class as a working example:

- You observe pairs $(x, y)_i$ for $i = 1, ..., n$.

- $x$ is in $R^5$, $y$ is a count variable

- The true generating function for $y$ is:

$$
y_i \sim \text{Poisson}(\exp(0.1x_1^2 + 0.2*x_2x_3 + 5\cos(x_4) + 2\log(\mid x_5 \mid)) + 30) \, .
$$

We will fit both a linear model and a simple feed forward neural network to the data then predict on a held out test dataset.

```python
%matplotlib inline
import matplotlib.pyplot as plt
import numpy as np
import os
import sys
```

```python
import torch
import torch.autograd as autograd
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader, TensorDataset
from torch.distributions import Poisson
```

```python
# Generate some data from our model
n_train = 1000
n_validate = 500
n_test = 5000
n_inputs = 5

# Sample the X's
X = np.random.normal(size=(n_train+n_validate+n_test, n_inputs))

# Sample the y's
y = np.random.poisson(np.exp(0.1*X[:,0]**2 + 0.2*X[:,1]*X[:,2] + 5 * np.cos(X[:,3]) + 2*np.log(np.abs(X[:,4]))) + 30)
```

```python
# Look at all the data
plt.hist(y, bins=100)
```

```python
# Zoom in on the smaller parts
plt.hist(y, bins=np.linspace(0,100, 101))
```

```python
'''Neural network to map from X to Poisson rate of y.'''
class PoissonNetworkModel(nn.Module):
    def __init__(self, n_inputs, is_nonlinear=True):
        # You have to call the super constructor to initialize models in pytorch
        super(PoissonNetworkModel, self).__init__()

        # How big is the input vector we are expecting?
        self.n_in = n_inputs

        # Define the internal function, before the link.
        # This is f(x) in Poisson(g(f(x)))
        if is_nonlinear:
          self.fc_in = nn.Sequential(
                  nn.Linear(self.n_in, 10),
                  nn.ReLU(),
                  nn.Linear(10, 1))
        else:
          self.fc_in = nn.Sequential(
                  nn.Linear(self.n_in, 1))

        self.link = nn.Softplus()
        # self.link = torch.exp # If you try this, it will blow up.

        # Set the weights to small values for numerical stability
        self.fc_in[0].weight.data = torch.Tensor(np.random.random(size=self.fc_in[0].weight.shape)*2e-1 - 1e-1)
        if is_nonlinear:
          self.fc_in[2].weight.data = torch.Tensor(np.random.random(size=self.fc_in[2].weight.shape)*2e-1 - 1e-1)

    def forward(self, x):
        # Get f(x), map through the softplus link, and return
        return self.link(self.fc_in(x))
```

```python

def fit_nn(X_train, y_train, # Training data to fit the model with
           X_validate, y_validate, # Validation data to know when to stop
            n_epochs=30, # Maximum number of epochs to run
            batch_size=100, # Number of samples to use in each minibatch
            lr=1e-5, # Learning rate (just set to a constant)
            ridge_penalty=0, # L2 penalty on the coefficients,
            nonlinear_model=True, # Use a linear or nonlinear model?
           ):
  # Get the number of inputs
  n_inputs = X_train.shape[1]

  # Create the pytorch datasets
  train_dataset = TensorDataset(torch.from_numpy(X_train).float(),
                                torch.from_numpy(y_train))
  val_dataset = TensorDataset(torch.from_numpy(X_validate).float(),
                              torch.from_numpy(y_validate))

  # Create data loader objects that handle creating mini-batches and shuffling
  train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
  val_loader = DataLoader(val_dataset, batch_size=batch_size, shuffle=False)

  # Initialize the model
  model = PoissonNetworkModel(n_inputs, is_nonlinear=nonlinear_model)

  # Setup the SGD method. Note: deep learning people call ridge penalties "weight decay"
  optimizer = optim.SGD(model.parameters(), lr=lr, weight_decay=ridge_penalty)

  # Track how the model changes over time
  train_losses = []
  validation_losses = []

  # Train the model
  for epoch in range(n_epochs):
    # Track the loss curves
    train_loss = torch.Tensor([0])

    # Take gradient steps using mini-batch SGD
    for batch_X, batch_y in train_loader:
      # Clear the pytorch cache of gradients. This is a technical detail.
      # You just call it every time you're about to start a new mini-batch.
      optimizer.zero_grad()

      # Get the predicted rates for each X sample in the mini-batch
      rate = model(batch_X)

      # Calculate the loss under a poisson likelihood
      dist = Poisson(rate.clip(1e-6)) # Clip for numerical stability
      loss = -dist.log_prob(batch_y).sum() # Add the individual sample losses

      # autograd backward pass (calculates derivatives using backprop).
      loss.backward()

      # Take SGD step (updates model weights).
      optimizer.step()

      # Update the overall loss for the entire training dataset
      train_loss += loss.data

    # Add the average loss to the overall progress curve
    train_losses.append(float(train_loss) / X_train.shape[0])

    # Now evaluate on held out data. First turn off training mode; this disables
    # the forward-mode caching which is expensive and unnecessary.
    model.eval()

    # Same as in training mode only no calculating gradients or updating weights.
    val_loss = torch.Tensor([0])
    for batch_X, batch_y in val_loader:
      rate = model(batch_X)
      dist = Poisson(rate.clip(1e-6, 1e6))
      loss = -dist.log_prob(batch_y).sum()
      val_loss += loss.data
    validation_losses.append(float(val_loss) / X_validate.shape[0])

  return model, train_losses, validation_losses
```

```python
X_train, X_validate, X_test = X[:n_train], X[n_train:n_train+n_validate], X[:-n_train+n_validate]
y_train, y_validate, y_test = y[:n_train], y[n_train:n_train+n_validate], y[:-n_train+n_validate]

# Let's fit it!
linear_model, linear_train_losses, linear_validation_losses = fit_nn(X_train, y_train,
                                                X_validate, y_validate,
                                                nonlinear_model=False)

nonlinear_model, nonlinear_train_losses, nonlinear_validation_losses = fit_nn(X_train, y_train,
                                                X_validate, y_validate,
                                                nonlinear_model=True)
```

```python
plt.plot(linear_train_losses, color='orange', label='linear train')
plt.plot(linear_validation_losses, color='blue', label='linear validation')
plt.plot(nonlinear_train_losses, color='orange', ls='--', label='nonlinear train')
plt.plot(nonlinear_validation_losses, color='blue', ls='--', label='nonlinear validation')
plt.legend(loc='upper right')
```
